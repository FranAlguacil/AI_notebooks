# Tema 4: Fundamentos de Patrones

## 4.1 Las Tres Dimensiones del Diseño Agéntico

Antes vimos estas dimensiones en el Tema 3. Ahora profundizamos para entender cómo guían el diseño de patrones.

### Recordatorio: Las 3 Dimensiones

```
┌────────────────────────────────────────────────────┐
│          LAS 3 DIMENSIONES DEL DISEÑO              │
└────────────────────────────────────────────────────┘

1️⃣ ARQUITECTURA (Estructura)
   Pregunta: ¿Cómo está CONSTRUIDO?
   │
   ├─ Single-Agent vs Multi-Agent
   ├─ Topología (P2P, Hierarchical)
   ├─ Componentes y boundaries
   └─ Puntos de integración
   
   Decisión: Design-time
   Cambia: Raramente

2️⃣ FLUJO/ORCHESTRATION (Comportamiento)
   Pregunta: ¿Cómo OPERA?
   │
   ├─ Secuencial (Chaining, Refinement)
   ├─ Condicional (Routing, Branching)
   ├─ Paralelo (Map-Reduce, Voting)
   └─ Iterativo (ReAct, Reflection)
   
   Decisión: Run-time
   Cambia: Dinámicamente

3️⃣ CAPACIDADES (Habilitadores)
   Pregunta: ¿Qué PUEDE HACER?
   │
   ├─ Tools (search, database, APIs)
   ├─ Memoria (working, episodic, semantic)
   ├─ LLM capabilities (reasoning, code)
   └─ Integraciones externas
   
   Decisión: Configuration-time
   Cambia: Por configuración
```

### Por qué importan estas dimensiones

**No son independientes** - cada una afecta a las otras:

- **Arquitectura determina** qué flujos son posibles
- **Capacidades limitan** qué arquitecturas son viables  
- **Flujo aprovecha** las capacidades disponibles

## 4.2 Relaciones Entre Dimensiones

### Arquitectura → Habilita → Flujos

**La arquitectura hace posible o imposible ciertos flujos.**

```
Single-Agent:
  ✅ PUEDE usar: ReAct, Reflection, Plan-and-Execute
  ❌ NO PUEDE usar: Debate (necesita múltiples agentes)
  ❌ NO PUEDE usar: Democratic voting

Multi-Agent Peer-to-Peer:
  ✅ PUEDE usar: Debate, Voting, Consensus
  ✅ PUEDE usar: ReAct (cada agente)
  ⚠️ COMPLEJO: Parallel execution (coordinación difícil)

Multi-Agent Hierarchical:
  ✅ PUEDE usar: Manager-Worker patterns
  ✅ PUEDE usar: Parallel execution (natural)
  ❌ NO PUEDE usar: Democratic debate (hay jerarquía)
```

### Capacidades → Permiten → Arquitecturas

**Las capacidades disponibles determinan qué arquitecturas tienen sentido.**

```
Sin tools:
  → Single-Agent suficiente
  → Multi-Agent es overkill

Con 100+ tools especializadas:
  → Multi-Agent
  → Single-Agent se vuelve confuso

Con capacidad de ejecución paralela:
  → Hierarchical architecture aprovecha mejor
  → Single-Agent desperdicia capacidad
```

### Flujo → Informa → Decisiones Arquitectónicas

**El flujo deseado puede forzar cambios arquitectónicos.**

```
Si necesitas Debate pattern:
  → DEBES usar Multi-Agent
  → Single-Agent no es opción

Si necesitas Parallel Map-Reduce:
  → Hierarchical es más natural
  → P2P es más complejo

Si necesitas ReAct simple:
  → Single-Agent es suficiente
  → Multi-Agent añade complejidad innecesaria
```

## 4.3 Matriz de Compatibilidad

### Tabla: Arquitectura × Flujo

Esta matriz te dice qué combinaciones tienen sentido.

| Flujo / Arquitectura | Single-Agent | Multi-Agent P2P | Multi-Agent Hierarchical |
|---|---|---|---|
| ReAct Loop | ✅ Ideal | ✅ Cada agente | ✅ Cada worker |
| Prompt Chaining | ✅ Ideal | ⚠️ Complejo | ✅ Pipeline |
| Reflection | ✅ Ideal | ✅ Multi-perspective | ✅ Por nivel |
| Plan-and-Execute | ✅ Ideal | ⚠️ Coordinación | ✅ Manager planifica |
| Routing | ✅ Ideal | ✅ Natural | ✅ Manager route |
| Map-Reduce (Parallel) | ⚠️ Secuencial | ⚠️ Coordinación | ✅ Ideal |
| Voting/Consensus | ❌ Imposible | ✅ Ideal | ⚠️ Solo workers |
| Debate | ❌ Imposible | ✅ Ideal | ⚠️ Solo workers |
| Manager-Worker | ❌ Imposible | ❌ No hay jerarquía | ✅ Ideal |

**📖 Interpretación:**  
- ✅ **Ideal fit:** Este flujo funciona naturalmente con esta arquitectura  
- ⚠️ **Posible pero complejo:** Funciona pero con overhead o limitaciones  
- ❌ **Contradictorio:** Este flujo requiere otra arquitectura

## 4.1 Patrones Secuenciales

### 4.1.1 Prompt Chaining

**Definición:** El output del paso N se convierte en el input del paso N+1.

```
Step 1: Input → LLM → Output₁
                        ↓
Step 2:         Input₂ = Output₁ → LLM → Output₂
                                          ↓
Step 3:                         Input₃ = Output₂ → LLM → Final Output
```

**Cuándo usar:**
- ✅ Transformaciones sucesivas bien definidas
- ✅ Cada paso tiene un objetivo claro
- ✅ Pasos son independientes (stateless)

**Ejemplo conceptual:**

````python
from google.colab import userdata
import os

# Fetch the API key from Colab Secrets
openai_api_key = userdata.get('OPENAI_API_KEY')

# Set it as an environment variable so the existing code can pick it up
os.environ['OPENAI_API_KEY'] = openai_api_key

print("OpenAI API key loaded from Colab Secrets and set as environment variable.")
````

````python
import os
import json
import re
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))


def llm_text(prompt: str, model: str = "gpt-4.1-mini") -> str:
    response = client.responses.create(
        model=model,
        input=prompt
    )
    return response.output_text.strip()


# Mini corpus local para la demo
KNOWLEDGE_BASE = [
    {
        "title": "Qué son los agentes de IA",
        "content": """
Los agentes de IA son sistemas que perciben información del entorno, razonan sobre ella
y ejecutan acciones para alcanzar un objetivo. A diferencia de un chatbot tradicional,
un agente puede planificar, usar herramientas y operar en varios pasos.
"""
    },
    {
        "title": "Capacidades típicas de un agente",
        "content": """
Las capacidades más comunes de un agente son el uso de tools, memoria de corto y largo plazo,
planificación, recuperación de información y ejecución de acciones externas como búsquedas,
consultas a bases de datos, envío de emails o ejecución de código.
"""
    },
    {
        "title": "Arquitecturas de agentes",
        "content": """
Un agente puede organizarse como single-agent o multi-agent. El single-agent centraliza
estado y razonamiento en un único modelo. El multi-agent distribuye tareas entre varios agentes
especializados, pero introduce costes de coordinación y latencia.
"""
    },
    {
        "title": "Patrones de flujo",
        "content": """
Entre los patrones más usados están ReAct, prompt chaining, reflection, plan-and-execute
y routing. ReAct alterna pensamiento, acción y observación. Prompt chaining encadena
salidas de un paso como entradas del siguiente.
"""
    },
    {
        "title": "Tendencias recientes",
        "content": """
En los últimos años han ganado importancia los agentes con tool use, RAG agéntico,
MCP para estandarizar integraciones, code execution en sandbox y sistemas multi-agente
para tareas complejas o especializadas.
"""
    }
]


def tokenize(text: str) -> set:
    text = text.lower()
    words = re.findall(r"\w+", text, flags=re.UNICODE)
    return set(words)


def retrieve_local_documents(query: str, top_k: int = 3) -> list:
    """
    Retrieval real sobre corpus local usando overlap de keywords.
    """
    query_tokens = tokenize(query)
    scored = []

    for doc in KNOWLEDGE_BASE:
        doc_tokens = tokenize(doc["title"] + " " + doc["content"])
        score = len(query_tokens.intersection(doc_tokens))
        scored.append((score, doc))

    scored.sort(key=lambda x: x[0], reverse=True)
    return [doc for score, doc in scored[:top_k] if score > 0]


def prompt_chaining_example(user_query: str, model: str = "gpt-4.1-mini") -> str:

    print("⛓️ PROMPT CHAINING\n")

    try:
        # Step 1
        print("Step 1: Extraer keywords para búsqueda")
        prompt_1 = f"""
        Extrae keywords útiles para buscar información relevante.

        Consulta:
        {user_query}

        Devuelve solo una línea con keywords separadas por espacios.
        """
        output_1 = llm_text(prompt_1, model=model)
        print(f"  Output₁: '{output_1}'\n")

        # Step 2
        print("Step 2: Buscar en corpus local")
        output_2 = retrieve_local_documents(output_1, top_k=3)
        print("  Output₂:")
        print(json.dumps(output_2, indent=2, ensure_ascii=False), "\n")

        if not output_2:
            return "No se encontraron documentos relevantes en el corpus local."

        # Step 3
        print("Step 3: Resumir resultados")
        prompt_3 = f"""
        Resume estos documentos para responder a la consulta del usuario.

        Consulta original:
        {user_query}

        Documentos recuperados:
        {json.dumps(output_2, indent=2, ensure_ascii=False)}

        Haz un resumen breve y claro en un párrafo.
        """
        output_3 = llm_text(prompt_3, model=model)
        print(f"  Output₃: '{output_3}'\n")

        # Step 4
        print("Step 4: Formatear respuesta")
        prompt_4 = f"""
        Convierte este resumen en bullets claros y útiles.

        Resumen:
        {output_3}

        Devuelve entre 3 y 5 bullets.
      """
        final = llm_text(prompt_4, model=model)
        print(f"  Final:\n{final}\n")

        return final

    except Exception as e:
        return f"Error en la cadena: {e}"


# Ejecutar
resultado = prompt_chaining_example("Cuéntame sobre agentes de IA")
print("Resultado final:\n")
print(resultado)
````

### 4.1.2 Sequential Refinement

**Definición:** Aplicar el mismo proceso iterativamente para mejorar el resultado.

```
Input → Process → Output₁
         ↓
      Output₁ → Process → Output₂ (mejor)
                 ↓
              Output₂ → Process → Output₃ (aún mejor)
```

**Diferencia con Chaining:**
- Chaining: Cada paso es DIFERENTE (extract → search → summarize)
- Refinement: El MISMO paso se repite (refine → refine → refine)

**Cuándo usar:**
- ✅ Mejora iterativa posible
- ✅ Refinamiento progresivo beneficioso
- ✅ Tienes stopping criteria claro

**Stopping criteria:**
1. Quality threshold alcanzado
2. Max iterations
3. No más mejoras detectables

## 4.2 Patrones Condicionales

### 4.2.1 Routing

**Concepto:** Dirigir la request al handler apropiado basado en algún criterio.

#### A. Intent-based Routing

**Clasificar el intent y dirigir al handler apropiado.**

```
User Query
    ↓
[Classifier]
    ↓
  Intent?
    ├─→ "technical_support" → Technical Handler
    ├─→ "billing" → Billing Handler
    └─→ "general" → General Handler
```

**Accuracy considerations:**
- Classifier debe ser muy preciso
- Routing incorrecto = mala UX
- Fallback para casos ambiguos

#### B. Semantic Routing (RAG)

**Usar embedding similarity para dirigir a knowledge base apropiada.**

```
Query: "How do I reset my password?"
    ↓
[Embed query]
    ↓
[Compare with knowledge bases]
    ├─ KB1 (Product Docs): 0.92 similarity ✅
    ├─ KB2 (Legal Docs): 0.23 similarity
    └─ KB3 (Marketing): 0.15 similarity
    ↓
Route to KB1
```

#### C. Model Routing (Cost Optimization)

**Estimate complexity y route a modelo apropiado.**

```
Query
    ↓
[Complexity Estimator]
    ↓
  Complexity?
    ├─→ Simple → GPT-3.5 (barato, rápido)
    ├─→ Medium → GPT-4 Turbo
    └─→ Complex → GPT-4o (caro, capaz)
```

**Trade-off:** Costo vs Performance

````python
# ============================================================
# A. INTENT-BASED ROUTING
# ============================================================

class IntentBasedRouter:
    """
    Router que clasifica el intent y dirige al handler apropiado
    """

    def __init__(self, llm):
        self.llm = llm
        self.handlers = {
            'technical_support': self.handle_technical,
            'billing': self.handle_billing,
            'general': self.handle_general
        }

    def route(self, user_query: str) -> dict:
        """
        Clasifica intent y route al handler apropiado
        """
        # Paso 1: Clasificar intent
        intent = self.classify_intent(user_query)

        print(f"📋 Query: {user_query}")
        print(f"🎯 Intent detectado: {intent['category']} (confidence: {intent['confidence']})")

        # Paso 2: Verificar confidence
        if intent['confidence'] < 0.7:
            print("⚠️ Baja confidence - usando fallback")
            return self.handle_ambiguous(user_query, intent)

        # Paso 3: Route al handler apropiado
        handler = self.handlers.get(intent['category'], self.handle_general)
        response = handler(user_query)

        return {
            'intent': intent['category'],
            'confidence': intent['confidence'],
            'response': response
        }

    def classify_intent(self, query: str) -> dict:
        """
        Clasifica el intent usando LLM
        """
        prompt = f"""
        Clasifica el siguiente query en UNA de estas categorías:
        - technical_support: problemas técnicos, bugs, errores
        - billing: pagos, facturas, precios, suscripciones
        - general: preguntas generales, información

        Query: "{query}"

        Responde en JSON:
        {{
            "category": "...",
            "confidence": 0.95,
            "reasoning": "..."
        }}
        """

        # Simular respuesta del LLM
        # En producción: response = self.llm.generate(prompt)

        # Simulacion
        if "bug" in query.lower() or "error" in query.lower():
            return {
                "category": "technical_support",
                "confidence": 0.95,
                "reasoning": "Menciona error/bug"
            }
        elif "invoice" in query.lower() or "payment" in query.lower():
            return {
                "category": "billing",
                "confidence": 0.90,
                "reasoning": "Menciona invoice/payment"
            }
        else:
            return {
                "category": "general",
                "confidence": 0.75,
                "reasoning": "Query genérica"
            }

    def handle_technical(self, query: str) -> str:
        print("🔧 Routing a: Technical Support Handler")
        return f"[TECHNICAL] Analizando problema técnico: {query}"

    def handle_billing(self, query: str) -> str:
        print("💰 Routing a: Billing Handler")
        return f"[BILLING] Procesando consulta de facturación: {query}"

    def handle_general(self, query: str) -> str:
        print("💬 Routing a: General Handler")
        return f"[GENERAL] Respondiendo consulta general: {query}"

    def handle_ambiguous(self, query: str, intent: dict) -> dict:
        """
        Fallback para casos ambiguos
        """
        print(f"❓ Query ambiguo - pidiendo clarificación")
        return {
            'intent': 'ambiguous',
            'confidence': intent['confidence'],
            'response': f"No estoy seguro si preguntas sobre {intent['category']}. ¿Podrías clarificar?"
        }


# Ejemplo de uso
print("="*60)
print("EJEMPLO: INTENT-BASED ROUTING")
print("="*60)

router = IntentBasedRouter(llm=None)

# Test 1: Technical
result1 = router.route("I'm getting an error 500 in the app")
print(f"✅ Respuesta: {result1['response']}\n")

# Test 2: Billing
result2 = router.route("Where can I download my invoice?")
print(f"✅ Respuesta: {result2['response']}\n")

# Test 3: General
result3 = router.route("What are your business hours?")
print(f"✅ Respuesta: {result3['response']}\n")
````

````python
# ============================================================
# B. SEMANTIC ROUTING (RAG)
# ============================================================

import numpy as np
from typing import List, Dict

class SemanticRouter:
    """
    Router basado en embedding similarity para RAG
    """

    def __init__(self):
        # Simular knowledge bases con embeddings
        self.knowledge_bases = {
            'product_docs': {
                'description': 'Product documentation and how-to guides',
                'embedding': np.array([0.8, 0.1, 0.2, 0.3, 0.5]),  # Simulado
                'keywords': ['password', 'login', 'setup', 'install', 'configure']
            },
            'legal_docs': {
                'description': 'Terms of service, privacy policy, legal',
                'embedding': np.array([0.2, 0.9, 0.1, 0.4, 0.3]),  # Simulado
                'keywords': ['terms', 'privacy', 'gdpr', 'legal', 'policy']
            },
            'marketing': {
                'description': 'Product features, pricing, comparisons',
                'embedding': np.array([0.3, 0.2, 0.9, 0.6, 0.4]),  # Simulado
                'keywords': ['price', 'features', 'compare', 'vs', 'plan']
            }
        }

    def route(self, query: str, threshold: float = 0.7) -> Dict:
        """
        Route query a la KB más relevante usando similarity
        """
        print(f"\n📋 Query: {query}")

        # Paso 1: Embed query
        query_embedding = self.embed_query(query)
        print(f"🔢 Query embedded: {query_embedding}")

        # Paso 2: Calcular similarity con cada KB
        similarities = {}
        for kb_name, kb_data in self.knowledge_bases.items():
            similarity = self.cosine_similarity(
                query_embedding,
                kb_data['embedding']
            )
            similarities[kb_name] = similarity

        # Paso 3: Mostrar scores
        print("\n📊 Similarity scores:")
        for kb_name, score in sorted(similarities.items(), key=lambda x: x[1], reverse=True):
            status = "✅" if score >= threshold else "❌"
            print(f"   {status} {kb_name}: {score:.2f}")

        # Paso 4: Elegir el mejor match
        best_kb = max(similarities.items(), key=lambda x: x[1])

        if best_kb[1] >= threshold:
            print(f"\n🎯 Routing a: {best_kb[0]} (similarity: {best_kb[1]:.2f})")
            return {
                'knowledge_base': best_kb[0],
                'similarity': best_kb[1],
                'description': self.knowledge_bases[best_kb[0]]['description']
            }
        else:
            print(f"\n⚠️ No hay KB con similarity >= {threshold}")
            return {
                'knowledge_base': 'general_fallback',
                'similarity': best_kb[1],
                'description': 'General knowledge base'
            }

    def embed_query(self, query: str) -> np.ndarray:
        """
        Simular embedding de query
        En producción: usar OpenAI embeddings o similar
        """
        # Simulación simple basada en keywords
        query_lower = query.lower()

        # Check keywords de cada KB
        product_score = sum(1 for kw in self.knowledge_bases['product_docs']['keywords'] if kw in query_lower)
        legal_score = sum(1 for kw in self.knowledge_bases['legal_docs']['keywords'] if kw in query_lower)
        marketing_score = sum(1 for kw in self.knowledge_bases['marketing']['keywords'] if kw in query_lower)

        # Crear embedding simple
        return np.array([
            0.8 if product_score > 0 else 0.2,
            0.8 if legal_score > 0 else 0.1,
            0.8 if marketing_score > 0 else 0.3,
            0.5,
            0.4
        ])

    def cosine_similarity(self, vec1: np.ndarray, vec2: np.ndarray) -> float:
        """
        Calcular cosine similarity entre dos vectores
        """
        dot_product = np.dot(vec1, vec2)
        norm_product = np.linalg.norm(vec1) * np.linalg.norm(vec2)
        return dot_product / norm_product if norm_product > 0 else 0.0


# Ejemplo de uso
print("\n" + "="*60)
print("EJEMPLO: SEMANTIC ROUTING (RAG)")
print("="*60)

semantic_router = SemanticRouter()

# Test 1: Product question
result1 = semantic_router.route("How do I reset my password?")
print(f"📁 KB seleccionada: {result1['knowledge_base']}")

# Test 2: Legal question
result2 = semantic_router.route("What is your privacy policy?")
print(f"📁 KB seleccionada: {result2['knowledge_base']}")

# Test 3: Marketing question
result3 = semantic_router.route("What are the pricing plans?")
print(f"📁 KB seleccionada: {result3['knowledge_base']}")
````

````python
# ============================================================
# C. MODEL ROUTING (COST OPTIMIZATION)
# ============================================================

class ModelRouter:
    """
    Router que selecciona el modelo óptimo según complejidad
    """

    def __init__(self):
        self.models = {
            'gpt-3.5-turbo': {
                'cost_per_1k': 0.0015,
                'capability_score': 7,
                'latency_ms': 500,
                'max_complexity': 3
            },
            'gpt-4-turbo': {
                'cost_per_1k': 0.01,
                'capability_score': 9,
                'latency_ms': 1000,
                'max_complexity': 7
            },
            'gpt-4o': {
                'cost_per_1k': 0.03,
                'capability_score': 10,
                'latency_ms': 1500,
                'max_complexity': 10
            }
        }

    def route(self, query: str, priority: str = 'balanced') -> Dict:
        """
        Route a modelo apropiado según complejidad y priority

        priority: 'cost', 'speed', 'quality', 'balanced'
        """
        print(f"\n📋 Query: {query}")
        print(f"⚙️ Priority: {priority}")

        # Paso 1: Estimar complejidad
        complexity = self.estimate_complexity(query)
        print(f"📊 Complexity score: {complexity}/10")

        # Paso 2: Seleccionar modelo según priority
        if priority == 'cost':
            model = self.select_cheapest_viable(complexity)
        elif priority == 'speed':
            model = self.select_fastest_viable(complexity)
        elif priority == 'quality':
            model = self.select_best_quality(complexity)
        else:  # balanced
            model = self.select_balanced(complexity)

        model_info = self.models[model]

        print(f"\n🎯 Modelo seleccionado: {model}")
        print(f"💰 Costo: ${model_info['cost_per_1k']}/1K tokens")
        print(f"⚡ Latencia: {model_info['latency_ms']}ms")
        print(f"🎓 Capability: {model_info['capability_score']}/10")

        return {
            'model': model,
            'complexity': complexity,
            'estimated_cost': model_info['cost_per_1k'],
            'estimated_latency': model_info['latency_ms']
        }

    def estimate_complexity(self, query: str) -> int:
        """
        Estimar complejidad de la query (1-10)
        """
        complexity = 1

        # Factores que aumentan complejidad
        if len(query.split()) > 50:
            complexity += 2

        if any(word in query.lower() for word in ['analyze', 'compare', 'explain', 'why']):
            complexity += 2

        if any(word in query.lower() for word in ['code', 'implement', 'debug', 'algorithm']):
            complexity += 3

        if '?' in query and query.count('?') > 1:
            complexity += 1

        return min(complexity, 10)

    def select_cheapest_viable(self, complexity: int) -> str:
        """
        Seleccionar el modelo más barato que pueda manejar la complejidad
        """
        for model, info in sorted(self.models.items(), key=lambda x: x[1]['cost_per_1k']):
            if info['max_complexity'] >= complexity:
                return model
        return 'gpt-4o'  # Fallback al más capaz

    def select_fastest_viable(self, complexity: int) -> str:
        """
        Seleccionar el modelo más rápido que pueda manejar la complejidad
        """
        for model, info in sorted(self.models.items(), key=lambda x: x[1]['latency_ms']):
            if info['max_complexity'] >= complexity:
                return model
        return 'gpt-4o'

    def select_best_quality(self, complexity: int) -> str:
        """
        Seleccionar el modelo de mejor calidad
        """
        return 'gpt-4o'  # Siempre el mejor

    def select_balanced(self, complexity: int) -> str:
        """
        Balance entre costo, velocidad y calidad
        """
        if complexity <= 3:
            return 'gpt-3.5-turbo'
        elif complexity <= 7:
            return 'gpt-4-turbo'
        else:
            return 'gpt-4o'


# Ejemplo de uso
print("\n" + "="*60)
print("EJEMPLO: MODEL ROUTING (COST OPTIMIZATION)")
print("="*60)

model_router = ModelRouter()

# Test 1: Simple query
print("\n--- Test 1: Simple query ---")
result1 = model_router.route(
    "What is the weather today?",
    priority='balanced'
)

# Test 2: Complex query
print("\n--- Test 2: Complex query ---")
result2 = model_router.route(
    "Write a Python implementation of a red-black tree with insertion, deletion, and balancing",
    priority='balanced'
)

# Test 3: Same query pero priorizando costo
print("\n--- Test 3: Cost priority ---")
result3 = model_router.route(
    "Write a Python implementation of a red-black tree with insertion, deletion, and balancing",
    priority='cost'
)

# Test 4: Same query pero priorizando calidad
print("\n--- Test 4: Quality priority ---")
result4 = model_router.route(
    "Write a Python implementation of a red-black tree with insertion, deletion, and balancing",
    priority='quality'
)
````

## 4.3 Patrones Paralelos

### 4.3.1 Parallel Sectioning (Map-Reduce)

**Map phase:** Dividir y procesar en paralelo  
**Reduce phase:** Combinar resultados

```
Large Document
    ↓
[Split into sections]
    ↓
  Section₁  Section₂  Section₃  (MAP)
     ↓         ↓         ↓
  Process  Process  Process    (PARALLEL)
     ↓         ↓         ↓
  Result₁  Result₂  Result₃
     └─────────┴─────────┘
              ↓
        [Combine]              (REDUCE)
              ↓
       Final Result
```

**Cuándo usar:**
- ✅ Documentos largos (>context window)
- ✅ Datasets grandes
- ✅ Procesamiento independiente posible

**Trade-offs:**
- ✅ Reduce latencia (paralelo)
- ✅ Maneja documentos grandes
- ❌ Context loss entre secciones
- ❌ Complejidad de combine phase

### 4.3.2 Voting (Consensus)

**Generar múltiples respuestas y elegir la mejor por voting.**

#### A. Majority Voting

```
Query
    ↓
[Generate N responses in parallel]
    ↓
Response₁  Response₂  Response₃
    ↓         ↓         ↓
[Vote for best response]
    ↓
Winner (most votes)
```

**Funciona para:** Discrete outputs (clasificación, opciones)

#### B. LLM-as-Judge

```
Query
    ↓
[Generate N candidates]
    ↓
Candidate₁  Candidate₂  Candidate₃
    └──────────┴──────────┘
              ↓
       [Judge LLM]
       "Evaluate and pick best"
              ↓
         Winner
```

**Funciona para:** Open-ended outputs (texto, código)

**Cuándo usar Voting:**
- ✅ Decisiones críticas (alta confiabilidad)
- ✅ Budget permite (Nx más caro)
- ✅ Latencia permite (paralelo pero más recursos)

**Cost consideration:** Si N=5, costo = 5x

````python
import asyncio
from openai import AsyncOpenAI
from collections import Counter
from typing import List, Dict
import os

class MajorityVotingSystem:
    """
    Sistema de voting para decisiones discretas usando LLM real
    """

    def __init__(self, n_candidates: int = 5):
        self.n_candidates = n_candidates
        self.client = AsyncOpenAI(api_key=os.getenv("OPENAI_API_KEY"))

    async def decide(self, query: str, options: List[str]) -> Dict:
        """
        Genera N respuestas con LLM y vota por la mejor
        """
        print(f"❓ Query: {query}")
        print(f"📋 Opciones: {options}")
        print(f"🗳️ Consultando {self.n_candidates} agentes LLM...\n")

        # FASE 1: Generar N respuestas en paralelo
        tasks = [
            self.ask_llm(i+1, query, options)
            for i in range(self.n_candidates)
        ]
        responses = await asyncio.gather(*tasks)

        # FASE 2: Votar
        winner = self.vote(responses)

        return {
            'winner': winner['option'],
            'votes': winner['votes'],
            'confidence': winner['votes'] / self.n_candidates,
            'all_responses': responses
        }

    async def ask_llm(self, agent_id: int, query: str, options: List[str]) -> str:
        """
        Consultar a un agente LLM
        """
        prompt = f"""Responde con UNA SOLA PALABRA de estas opciones: {', '.join(options)}

Pregunta: {query}

Responde solo con: {' o '.join(options)}"""

        response = await self.client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.7,  # Variabilidad entre agentes
            max_tokens=10
        )

        choice = response.choices[0].message.content.strip()
        print(f"  🤖 Agente {agent_id}: {choice}")

        return choice

    def vote(self, responses: List[str]) -> Dict:
        """
        Contar votos y determinar ganador
        """
        vote_counts = Counter(responses)

        print("\n📊 Resultados del voting:")
        for option, count in vote_counts.most_common():
            percentage = (count / len(responses)) * 100
            bar = "█" * count
            print(f"   {option}: {bar} ({count} votos, {percentage:.1f}%)")

        winner = vote_counts.most_common(1)[0]

        return {
            'option': winner[0],
            'votes': winner[1]
        }


# Ejemplo de uso
async def main():
    print("="*70)
    print("EJEMPLO: MAJORITY VOTING CON LLM REAL")
    print("="*70 + "\n")

    voting_system = MajorityVotingSystem(n_candidates=5)

    result = await voting_system.decide(
        query="Should we proceed with the product launch?",
        options=["Yes", "No", "Wait"]
    )

    print("\n" + "="*70)
    print("DECISIÓN FINAL:")
    print("="*70)
    print(f"🏆 Ganador: {result['winner']}")
    print(f"📊 Votos: {result['votes']}/{voting_system.n_candidates}")
    print(f"💯 Confianza: {result['confidence']*100:.1f}%")
    print(f"\n💰 Costo: {voting_system.n_candidates}x llamadas a LLM")

# Ejecutar
await main()
````

## 4.4 Patrones Iterativos

### 4.4.1 ReAct Loop

**Ya lo vimos en Tema 2, pero aquí va el recap:**

```
┌─→ THOUGHT: "Necesito buscar información"
│      ↓
│   ACTION: search_web("query")
│      ↓
│   OBSERVATION: "Resultados encontrados"
│      ↓
│   ¿Goal achieved?
│      ├─ No → loop ┘
│      └─ Sí → END
```

**Stop conditions:**
1. Goal achieved
2. Max iterations
3. No progress (stuck detection)

**Optimizaciones:**
- Early stopping si detectas loop
- Context windowing para manejar contexto largo
- Planning hints para guiar el razonamiento

### 4.4.2 Reflection Loop

**Generate → Reflect → Evaluate → Refine → Loop**

```
┌─→ GENERATE: Crear solución inicial
│      ↓
│   REFLECT: "¿Qué está mal? ¿Qué mejorar?"
│      ↓
│   EVALUATE: Score quality (1-10)
│      ↓
│   ¿Quality > threshold?
│      ├─ No → REFINE → loop ┘
│      └─ Sí → DONE
```

**Self-critique strategies:**

**A. Single agent reflection:**
```
Same agent:
  1. Generate
  2. Critique own work
  3. Refine based on critique
```

**B. Multi-agent (Actor-Critic):**
```
Actor: Generate
Critic: Evaluate
Actor: Refine based on feedback
```

**Trade-offs:**
- ✅ Mejora quality significativamente
- ❌ Costo alto (múltiples iteraciones)
- ❌ Latencia alta

### 4.4.3 Plan-and-Execute

**Planning phase:** Generar plan completo upfront  
**Execution phase:** Ejecutar cada paso del plan

```
PLANNING PHASE:
    Task → Analyze → Generate complete plan
    Plan: [Step 1, Step 2, Step 3, ...]
    
EXECUTION PHASE:
    For each step in plan:
        Execute step
        Verify result
        If failure → Re-plan
```

**ReAct vs Plan-and-Execute:**

| Aspecto | ReAct | Plan-and-Execute |
|---------|-------|------------------|
| Planning | Incremental (paso a paso) | Upfront (todo al inicio) |
| Flexibility | Alta (se adapta) | Media (puede re-plan) |
| Efficiency | Menos (más LLM calls) | Más (menos calls) |
| Best for | Tareas exploratorias | Tareas bien definidas |

**Cuándo usar cada uno:**
- **ReAct:** Cuando no sabes qué encontrarás (exploración)
- **Plan-and-Execute:** Cuando sabes los pasos (ejecución)

### 4.4.4 ReWOO (Reasoning Without Observation)

**Plan all actions upfront, execute sin intermediate reasoning.**

```
1. REASON: Plan ALL actions
   "Primero buscar X, luego analizar Y, finalmente Z"
   
2. EXECUTE: Run all actions (sin reasoning intermedio)
   Execute(action_1)
   Execute(action_2)
   Execute(action_3)
   
3. SYNTHESIZE: Combine results
```

**Efficient:** Menos LLM calls (solo 2: plan + synthesize)

**Cuándo usar:**
- ✅ Tareas bien definidas
- ✅ Pasos predecibles
- ✅ Cost optimization crítico

**Limitations:**
- ❌ Menos adaptable (no puede cambiar en medio)
- ❌ Errores se propagan (no hay early detection)

### 4.4.5 Evaluator-Optimizer

**Loop de mejora continua driven por métricas.**

```
┌─→ GENERATE: Create solution
│      ↓
│   EVALUATE: Measure metrics
│      (accuracy, latency, cost, etc.)
│      ↓
│   OPTIMIZE: Improve based on metrics
│      ↓
│   ¿Metrics acceptable?
│      ├─ No → loop ┘
│      └─ Sí → DONE
```

**Use cases:**
- Prompt optimization
- Hyperparameter tuning
- A/B testing automation

**Métrica-driven approach** es clave para sistemas en producción.


# 4.5 Patrones de Coordinación



## 4.5.1 Manager Coordination (Orchestrator-Worker)

**Ya lo vimos en Tema 3 como arquitectura Hierarchical. Aquí el patrón de coordinación:**

```
        MANAGER
           │
    1. Decompose task
           │
    2. Assign to workers
           │
    ┌──────┼──────┐
    ▼      ▼      ▼
Worker₁ Worker₂ Worker₃
    │      │      │
    3. Execute subtasks
    │      │      │
    └──────┴──────┘
           │
    4. Manager aggregates
```

**Challenges:**
- Manager como bottleneck
- Load balancing (distribuir trabajo equitativamente)

**Best practices:**
- Manager debe ser eficiente (no hacer trabajo pesado)
- Workers deben ser especializados
- Aggregation debe manejar resultados inconsistentes

## 4.5.2 Democratic Coordination (Peer-to-Peer)

**Agentes iguales sin jerarquía.**

### Consensus Mechanisms

#### A. Voting
```
Each agent votes
Majority wins
```

#### B. Debate Rounds
```
Round 1: Each agent proposes
Round 2: Agents critique each other
Round 3: Refined proposals
Round 4: Vote
```

#### C. Compromise
```
Agents negotiate
Find middle ground
Synthesize solution
```

**Challenges:**
- Coordinación compleja (sin líder)
- Evitar deadlocks (agentes no llegan a acuerdo)
- Costo alto (muchas rondas de comunicación)

## 4.5.3 Hierarchical Coordination (Multi-nivel)

**Múltiples niveles de managers.**

```
      Top Manager
           │
     ┌─────┴─────┐
     ▼           ▼
  Manager₁    Manager₂
     │           │
  ┌──┴──┐     ┌──┴──┐
  ▼     ▼     ▼     ▼
W₁.₁  W₁.₂  W₂.₁  W₂.₂
```

**Cuándo usar:**
- Tareas MUY complejas
- Múltiples sub-proyectos

**Challenges:**
- Overhead de coordinación alto
- Information loss entre niveles
- Difícil de debuggear

```
┌─→ ACTOR: Generate solution
│      ↓
│   CRITIC: Evaluate quality
│      "This is good, but improve X"
│      ↓
│   ACTOR: Refine based on feedback
│      ↓
│   ¿Quality acceptable?
│      ├─ No → loop ┘
│      └─ Sí → DONE
```

**Specialization benefits:**
- Actor: Optimizado para generación
- Critic: Optimizado para evaluación

**Cuándo usar:**
- Quality crítica
- Verificación necesaria
- Budget permite (2x agentes)
